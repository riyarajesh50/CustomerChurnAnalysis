```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:/CustomerChurnAnalysis/data/cleaned_telco_churn.csv")
```

```python
df.groupby('Contract')['Churn'].value_counts().unstack().plot(kind='bar')
plt.title("Churn by Contract")
plt.savefig("D:/CustomerChurnAnalysis/visuals/churn_by_contract.png")
plt.show()
```

```python
df.boxplot(column='tenure', by='Churn')
plt.savefig("D:/CustomerChurnAnalysis/visuals/tenure_vs_churn.png")
plt.show()
```
