```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv("D:/CustomerChurnAnalysis/data/cleaned_telco_churn.csv")
df = pd.get_dummies(df, drop_first=True)

X = df.drop('Churn_Yes', axis=1)
y = df['Churn_Yes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt

importance = model.feature_importances_
features = X.columns

feat_imp = pd.Series(importance, index=features).sort_values(ascending=False)[:10]

feat_imp.plot(kind='bar')
plt.title("Top Churn Drivers")
plt.savefig("D:/CustomerChurnAnalysis/visuals/feature_importance.png")
plt.show()
```
